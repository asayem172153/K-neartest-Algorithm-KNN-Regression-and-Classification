# Step 01: Regression

```python
import pandas as pd
df = pd.read_csv('/kaggle/input/weight-height/weight-height.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df4LR = df.copy()
df4KNNR = df.copy()
df4KNNC = df.copy()
```

2. Separate x(Gender, Height) and y (y=Weight)
3. Train = 70%, Test = 30%
4. Apply Linear Regression
5. Evaluate the Model (Testing and training Accuracy, MSE for testing)
6. Apply KNN Regressor: https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsRegressor.html

Work from here..

```python
x = df4LR.drop('Weight', axis=1)
x.head()
```

```python
for column in x.columns:
    if x[column].dtype == 'object':
        dummies = pd.get_dummies(x[column],drop_first=True, prefix=column)
        x.drop(x[[column]], axis=1,inplace=True)
        x=pd.concat([x,dummies],axis=1)
```

```python
x.head()
```

```python
y = df4LR[['Weight']]
y.head()
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain,xtest,ytrain,ytest = tts(x,y,test_size=0.30,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
LR = LinearRegression()
```

```python
LR.fit(xtrain,ytrain)
```

# **Accuracy & MSE (Linear Regression)**

```python
#testing Accuracy
testAcc = LR.score(xtest,ytest)
testAcc
```

```python
#training Accuracy
trainAcc = LR.score(xtrain,ytrain)
trainAcc
```

```python
#MSE for testing
from sklearn.metrics import mean_squared_error
testMSE = mean_squared_error(ytest, LR.predict(xtest))
testMSE
```

# KNN Regression 

```python
x = df4KNNR.drop('Weight', axis=1)
x.head()
```

```python
for column in x.columns:
    if x[column].dtype == 'object':
        dummies = pd.get_dummies(x[column],drop_first=True, prefix=column)
        x.drop(x[[column]], axis=1,inplace=True)
        x=pd.concat([x,dummies],axis=1)
```

```python
x.head()
```

```python
y = df4KNNR[['Weight']]
y.head()
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain,xtest,ytrain,ytest = tts(x,y,test_size=0.30,random_state=1)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
KNN_RG = KNeighborsRegressor(3)
```

```python
KNN_RG.fit(xtrain,ytrain)
```

# **Accuracy & MSE (KNN Regression)**

```python
from sklearn.metrics import mean_squared_error
```

```python
#testing accuracy
testing_accuracy = KNN_RG.score(xtest,ytest)
testing_accuracy
```

```python
#training accuracy
training_accuracy = KNN_RG.score(xtrain,ytrain)
training_accuracy
```

```python
#MSE
testing_MSE=mean_squared_error(ytest,KNN_RG.predict(xtest))
testing_MSE
```

# comparison LR vs KNN Reg

```python
print('\t\t\t','Linear Regression','\tvs\t','KNN Regression\n')
print('Testing Accuracy','\t',testAcc,'\t\t',testing_accuracy,'\n')
print('Training Accuracy','\t',trainAcc,'\t\t',training_accuracy,'\n')
print('MSE\t\t\t',testMSE,'\t\t',testing_MSE,'\n')
```

  **After evaluating all the values we can see Linear Regression model performs better in this case execpt training Accuracy**
